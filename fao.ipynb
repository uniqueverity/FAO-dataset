```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
year_columns = fao[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
correlation_matrix = year_columns.corr()
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap for Years (Y2014 to Y2018)')
plt.show()
```

```python
sns.pairplot(year_columns, diag_kind='kde', markers='o', palette='coolwarm')
plt.suptitle('Pair Plot of Years (Y2014 to Y2018)')
plt.show()
```
